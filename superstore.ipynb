```python
#converting dates
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
#creating order year
df["order_year"] = df["Order Date"].dt.year
```

```python
#creating order month
df["order_month"] = df["Order Date"].dt.month_name()
```

```python
#create shipping dates
df["shipping_days"] = (
    df["Ship Date"] - df["Order Date"]
).dt.days
```

```python
import pandas as pd

df['Order Date'] = pd.to_datetime(
    df['Order Date'],
    format='%d-%m-%Y'
)

df['Ship Date'] = pd.to_datetime(
    df['Ship Date'],
    format='%d-%m-%Y'
)
```

```python
#Who are the top customers by sales?
df.groupby('Customer Name')['Sales'].sum()\
.sort_values(ascending=False)\
.head(10)
```

```python
#Which products generate the most revenue?
df.groupby('Product Name')['Sales'].sum()\
.sort_values(ascending=False)\
.head(10)
```

```python
#Do discounts affect profit?
df.groupby('Discount')['Profit'].mean()
```

```python
#Which products generate losses?
df.groupby('Product Name')['Profit'].sum()\
.sort_values()\
.head(10)
```

```python
#Which states generate the most sales?
df.groupby('State')['Sales'].sum()\
.sort_values(ascending=False)\
.head(10)
```

```python
#Which cities are most profitable?
df.groupby('City')['Profit'].sum()\
.sort_values(ascending=False)\
.head(10)
```

```python
#customers.csv
customers = df[
[
'Customer ID',
'Customer Name',
'Segment'
]
].drop_duplicates()
```

```python
#products.csv
products = df[
[
'Product ID',
'Category',
'Sub/Category',
'Product Name'
]
].drop_duplicates()
```
